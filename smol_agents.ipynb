# Smolagents Demo Notebook

This notebook demonstrates how to use Smolagents to create AI agents that can generate and execute code.

```python
import os
from dotenv import load_dotenv
from smolagents import CodeAgent, HfApiModel

# Load environment variables
load_dotenv()
```

## Initialize the Agent

We'll create a CodeAgent using a Hugging Face model. Make sure you have your Hugging Face API token in the `.env` file.

```python
agent = CodeAgent(
    tools=[],  # Required parameter, even if empty
    model=HfApiModel(
        token=os.getenv("HUGGINGFACE_API_TOKEN"),
        model_id="Qwen/Qwen2.5-Coder-32B-Instruct"
    )
)
```

## Example 1: Calculator Function

Let's create a simple calculator function that can perform basic arithmetic operations.

```python
task = """
Create a Python function that can perform basic arithmetic operations (add, subtract, multiply, divide).
The function should take three parameters: operation (string), num1 (float), and num2 (float).
Return the result of the operation.
"""

result = agent.run(task)
print("Generated code:")
print(result)
```

## Example 2: List Operations

Let's create a function that performs various list operations.

```python
task = """
Create a Python function that can perform the following list operations:
1. Remove duplicates while preserving order
2. Find the second largest number
3. Rotate the list by k positions

The function should take a list of numbers and an operation type as parameters.
"""

result = agent.run(task)
print("Generated code:")
print(result)
```

## Example 3: String Manipulation

Let's create a function for advanced string manipulation.

```python
task = """
Create a Python function that can perform the following string operations:
1. Reverse words in a sentence while preserving punctuation
2. Check if a string is a palindrome (ignoring spaces and punctuation)
3. Convert a string to title case while handling special cases

The function should take a string and an operation type as parameters.
"""

result = agent.run(task)
print("Generated code:")
print(result)
```
